```python
def download_captcha(captcha_url, agent):#保存图片
    captcha_name = re.findall("id=(.*?):", captcha_url)
    filename = "douban_%s.jpg" % (str(captcha_name[0]))
    with open(filename, 'wb') as f:
        # 以二进制写入的模式在本地构建新文件
        header = {
            'User-Agent': agent,
            'Referer': captcha_url
        }
        f.write(requests.get(captcha_url, headers=header).content)
        print(filename)
    return filename

def recognition_captcha(filename):#识别
    APP_ID = '17366818'
    API_KEY = 'or3FvGGA5wWKS7QnhngLZnag'
    SECRET_KEY = 'YD9VSadkgYuXCQc34SnYfBTLI840lOh3'
    
    client = AipOcr(APP_ID, API_KEY, SECRET_KEY)

    # 读取图片
    def get_file_content(file_path):
        with open(file_path, 'rb') as fp:
            return fp.read()
    
    options = {}
    options["language_type"] = "ENG"

    image = get_file_content(filename)
    result = json.dumps(client.basicGeneral(image,options))
    
    result=json.loads(result)

    if 'words_result' in result:
        return (''.join([i['words'] for i in result['words_result']]))
```
